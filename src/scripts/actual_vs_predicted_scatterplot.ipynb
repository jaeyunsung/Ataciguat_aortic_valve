```r
library("ggplot2")
```

```r
#read in the predicted results
results = read.csv("../output/neural_net_classifier_results_428.csv")
```

```r
#read in the actual results for delta area
delta_area_df = read.csv("../data/delta_area_for_regression.csv")
```

```r
#combine for plotting
all_data = cbind(delta_area_df,results)
```

```r
#make treatment a factor for plotting
all_data$Treatment_factor = factor(all_data$Treatment.Group)
```

```r
scatter_plot1 = ggplot(all_data,(aes(x=DELTA_AREA,y=predict)))+geom_point()+ geom_abline(intercept=0, slope=1)+ylim(-0.5,0.5)+xlim(-0.5,0.5)+geom_point(aes(colour=Treatment_factor))+
                scale_fill_manual(values=c("#2836CC", "#FFBA01"))+ theme(aspect.ratio = 1)
print(scatter_plot1)

#to save output, uncomment the following three lines:
pdf(file="../output/actual_vs_predicted_Delta_area_plot.pdf")
print(scatter_plot1)
dev.off()
```
