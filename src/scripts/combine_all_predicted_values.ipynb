```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#now read in the results from cross validation 
#read in the results for predicted and actual
prediction_results1 = pd.read_csv("../output/actual_vs_predicted/predicted_value1.csv")
for i in range(2,24):
    prediction_file = "../output/actual_vs_predicted/predicted_value%s.csv" % (i)
    prediction_results2 = pd.read_csv(prediction_file)
    prediction_results1 = pd.concat([prediction_results1, prediction_results2])

prediction_results1
```

```python
#read in the actual values to calculate the pearson correlation coefficient
actual_DELTA_AREA = pd.read_csv("../data/delta_area_for_regression.csv")
```

```python
print(stats.pearsonr(prediction_results1["predict"],actual_DELTA_AREA["DELTA_AREA"]))
```
